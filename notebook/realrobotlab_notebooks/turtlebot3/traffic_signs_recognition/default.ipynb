# Real Robot Lab Simulation

Welcome to the simulation of the Real Robot Lab. This is an exact representation of what you see in the live stream at the home page of the construct:

https://app.theconstructsim.com/#/Home

Since the slot allowed for connections for the Real Robot Lab are 30 minutes, you will probably want to use this simulation before and after your slot to improve and test your programs. So don't worry if you feel like you didn't do everything you wanted with the turltlebot at the Real Robot Lab, this is why this simulation is here!

## How to Launch Simulation

To start the simulation, first we'll need to source our workspace:

- Open a shell by clicking the Tools dropdown menu in the top left corner:



- Copy and paste the following terminal commands in your shell:

```python
cd simulation_ws
```

```python
source devel/setup.bash
```

- This sets up our environment and now we can go ahead and launch the simulation. Run the following command on a shell:

```python
roslaunch realrobotlab main.launch
```

And that's it! You should be able to see the simulation and control everything as if it was the real robot if you go to the Gazebo button in the Tools dropdown menu on the top left side of your screen:



You can also launch the simulation by skipping the previous shell command and just going to the Simulations dropdown menu on the top left side of your screen and selecting the blue button "Choose launch file." Look for the file main.launch under the realrobotlab package:




You should see this simulation now:
